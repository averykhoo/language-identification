```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
with open('ucd.all.flat.xml', 'rt', encoding='utf8') as f:
    xml_data = f.read()
```

```python
soup = BeautifulSoup(xml_data, 'lxml-xml')
```

```python
chars = soup.find_all('char')
len(chars)
```

```python
sc = dict()
scx = dict()
blk = dict()

sc_scx = dict()
sc_blk = dict()
scx_blk = dict()

weird_sc = set()

for char in chars:
    if char.get('cp') is not None:
        
        # only traditional chars have a simplified variant specified, and vice versa
        if char.get('kSimplifiedVariant') is not None:
            scx.setdefault('hant', set()).add(int(char['cp'], 16))
        if char.get('kTraditionalVariant') is not None:
            scx.setdefault('hans', set()).add(int(char['cp'], 16))
        
        # jamo is a subset of hangul, based on the block name
        if 'jamo' in char['blk'].lower():
            scx.setdefault('jamo', set()).add(int(char['cp'], 16))
        
        # main script for this char
        sc.setdefault(char['sc'], set()).add(int(char['cp'], 16))
        
        # other scripts that use this char
        for x in char['scx'].split():
            scx.setdefault(x, set()).add(int(char['cp'], 16))
            
            if char['sc'] not in char['scx']: # or char['sc'][0] == 'Z', char
                weird_sc.add(char['sc'])
            
            sc_scx.setdefault(char['sc'], set()).add(x)
            if char['sc'] == 'Zyyy' and x == 'Latn':
                print(char['gc'])
            scx_blk.setdefault(x, set()).add(char['blk'])
        sc_blk.setdefault(char['sc'], set()).add(char['blk'])
        
        blk.setdefault(char['blk'], set()).add(int(char['cp'], 16))
        
        
    else:
        assert char.get('first-cp') is not None, char
        assert char.get('last-cp') is not None, char
```
