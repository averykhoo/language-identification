```python
_1_to_2 = {_1 if pd.notna(_1) else _2:_2 for _, (_1, _2) in df[['639-1', '639-2']].iterrows()}
```

```python
from bs4 import BeautifulSoup
```

```python
with open('supplementalData.xml', 'r', encoding='utf8') as f:
    soup = BeautifulSoup(f.read(), 'lxml-xml')
```

```python
lang_scripts = dict()
for lang in soup.find_all('language'):
    if lang.get('scripts') is not None:
        lang_scripts.setdefault(_1_to_2.get(lang['type'], lang['type']), set()).update(lang['scripts'].split())
lang_scripts['eng']
```

```python
import csv
```

```python
rows = []
with open('iso_639-2.csv', 'rt', encoding='utf8', newline='') as f:
    c = csv.reader(f)
    for row in c:
        rows.append(row)
with open('iso639-2.tab', 'rt', encoding='utf8', newline='') as f:
    c = csv.reader(f, delimiter='\t')
    for row in c:
        rows.append(row)
```

```python
with open('iso639-2.csv', 'wt', encoding='utf8', newline='') as f:
    c = csv.writer(f)
    for row in rows:
        c.writerow(row)
```
