```python
import json
import pandas as pd
```

```python
df = pd.read_csv('iso639/iso639_2.csv', encoding='utf8')
df
```

```python
with open('iso15924/char_scripts_4.json') as f:
    sc = json.load(f)
sc.keys()
```

```python
pd.set_option('display.max_rows', 500)
```

```python
df[df['Scripts'].isna()]
```

```python
script_langs = dict()
for i, row in df.iterrows():
    if pd.notna(row['Scripts']):
        for s in row['Scripts'].split():
            assert s in sc, s
            script_langs.setdefault(s, set()).add(row['639-2'])
```

```python
script_langs.setdefault('Hani', set()).update(script_langs.get('Hanb', set()))
script_langs.setdefault('Bopo', set()).update(script_langs.get('Hanb', set()))
script_langs.setdefault('Hira', set()).update(script_langs.get('Hrkt', set()))
script_langs.setdefault('Kana', set()).update(script_langs.get('Hrkt', set()))
script_langs.setdefault('Hira', set()).update(script_langs.get('Jpan', set()))
script_langs.setdefault('Kana', set()).update(script_langs.get('Jpan', set()))
script_langs.setdefault('Hani', set()).update(script_langs.get('Jpan', set()))
script_langs.setdefault('Hrkt', set()).update(script_langs.get('Jpan', set()))
script_langs.setdefault('Hang', set()).update(script_langs.get('Kore', set()))
script_langs.setdefault('Hani', set()).update(script_langs.get('Kore', set()))
```

```python
script_langs
```

```python
data = dict()
for script in sorted(sc.keys()):
    data[script] = {
        'langs': sorted(script_langs.get(script, set())),
        'chars': sc[script]
    }
data
```

```python
[scr for scr, x in data.items() if not x['langs'] and len(x['chars']) > 0]
```

```python
data['Yezi']['langs'].append('kmr')
data['Jamo']['langs'].append('kor')
```

```python
with open('language_identification/scripts.json', 'w', encoding='ascii') as f:
    json.dump(data, f, indent=4)
```
