```python
from tmp import find_words
import pandas as pd
from pathlib import Path
import json
from collections import Counter
```

```python
word_langs = dict()
for path in Path('.').glob('*/_*.txt'):
    print(path)
    lang = path.stem[1:]

    with path.open(encoding='utf8') as f:
        for word in f.read().split():
            word_langs.setdefault(word, Counter()).update([lang])
len(word_langs)
```

```python
with open('stopwords-iso.json', encoding='utf8') as f:
    data = json.load(f)
    for lang, words in data.items():
        for word in find_words(' '.join(words)):
            word_langs.setdefault(word, Counter()).update([lang])
```

```python
uniq_words = set(word for word, langs in word_langs.items() if len(langs) == 1)
len(uniq_words)
```

```python
lang_words = dict()
for word in uniq_words:
    for lang, count in word_langs[word].items():
        if count > 1:
            lang_words.setdefault(lang, set()).add(word)
len(lang_words)
```

```python
for lang, words in lang_words.items():
    print(lang)
    with open(f'merged/_{lang}.txt', 'w', encoding='utf8') as f:
        for word in sorted(words):
            f.write(word + '\n')
```
