```python
from pathlib import Path
import csv
from collections import Counter
```

```python
len(list(Path('ngrams').glob('*1gram*')))
```

```python
counter = Counter()
words = set()
for path in Path('ngrams').glob('*1gram*'):
    print(path)
    with path.open('r', encoding='utf8', newline='') as f:
        c = csv.reader(f)
        for word, str_count in c:
            words.add(word)
        counter.update(words)
        words.clear()
```

```python
counter.most_common(1000)
```

```python
len(counter)
```

```python
with open('common_words.csv', 'w', encoding='utf8', newline='') as f:
    c = csv.writer(f)
    for word in counter:
        count = counter[word]
        if count >= 3:
            c.writerow([word, count])
```

```python
counter2 = Counter()
for word, count in counter.items():
    if count >= 5:
        counter2[word] = count
```

```python
len(counter2)
```

```python
with open('common_words.csv', 'w', encoding='utf8', newline='') as f:
    c = csv.writer(f)
    for word, count in counter2.most_common():
        c.writerow([word, count])
```
