```python
import csv

import tmp
```

```python
files = dict()
with open('MASTER.LINGUISTIC.1GRAM.big.csv', 'r' ,encoding='utf8', newline='') as f:
    
    c = csv.reader(f)
    for i, row in enumerate(c):
        if (i + 1) % 1000000 == 0:
            print(i + 1)
            
        # parse row
        _lang = tmp.lang_codes[row[0]]
        _words = list(tmp.find_words(row[1]))
        _count = int(row[2])
        
        # unwanted lang
        if _lang is None:
            continue
        
        # find output file
        if row[0] in files:
            _f, _c = files[row[0]]
        else:
            _f = open(f'ngrams/word-1gram-{_lang}--{row[0]}.big.csv', 'w', encoding='utf8', newline='')
            _c = csv.writer(_f)
            files[row[0]] = (_f, _c)
            
        # write fixed output to file
        for word in _words:
            _c.writerow([word.casefold(), _count])
```

```python
for _f, _c in files.values():
    _f.close()
```

```python
files = dict()
with open('MASTER.LINGUISTIC.2GRAM.big.csv', 'r' ,encoding='utf8', newline='') as f:
    
    c = csv.reader(f)
    for i, row in enumerate(c):
        if (i + 1) % 1000000 == 0:
            print(i + 1)
            
        # parse row
        _lang = tmp.lang_codes[row[0]]
        _words = list(tmp.find_words(row[1]))
        _bigrams = [' '.join(bigram) for bigram in tmp.ngrams(_words, 2)]
        _count = int(row[2])
        
        # unwanted lang
        if _lang is None:
            continue
        
        # find output file
        if row[0] in files:
            _f, _c = files[row[0]]
        else:
            _f = open(f'ngrams/word-2gram-{_lang}--{row[0]}.big.csv', 'w', encoding='utf8', newline='')
            _c = csv.writer(_f)
            files[row[0]] = (_f, _c)
            
        # write fixed output to file
        for word in _bigrams:
            _c.writerow([word.casefold(), _count])
```

```python
for _f, _c in files.values():
    _f.close()
```
